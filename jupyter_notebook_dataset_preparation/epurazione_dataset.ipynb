```python
_id = 1
QoE = pd.DataFrame()
dic = {}
tr = complete_dataset.set_index(" id ")
while _id<401:
    q = tr["QoE"][_id][:1][_id]
    dic[" id "] = _id
    dic["QoE"] = q
    QoE = QoE.append(dic,ignore_index=True)
    _id = _id+1
```
